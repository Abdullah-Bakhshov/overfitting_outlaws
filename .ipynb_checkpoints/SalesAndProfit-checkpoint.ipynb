## Reading CSV file

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("superstore.csv")

# Add a "Gross Profit" column
df['Gross Profit'] = df['Sales'] + df['Profit']

df
```

```python
# Getting unique categoreis, sub categories and product names
categories = df["Category"].unique()

print(categories)
```

```python
# Finding the cumliative sales and profit per category
cat_1 = df[df["Category"]==categories[0]]
cat_2 = df[df["Category"]==categories[1]]
cat_3 = df[df["Category"]==categories[2]]

all_cats = [cat_1,cat_2,cat_3] 
sales = []
profits = []
gross_profits = []

for x in all_cats:
    total_sales = x["Sales"].sum()
    sales.append(total_sales)
    
    total_profit = x["Profit"].sum()
    profits.append(total_profit)  

    gross = x["Gross Profit"].sum()
    gross_profits.append(gross)

# Creating a new table wth just categories
category_df = pd.DataFrame({
    "Category": categories,
    "Sales":sales,
    "Gross":gross_profits,
    "Profit":profits,
})

category_df
```

```python
# Plotting the cumulative sales and gross profit per category on pie charts

# Create a figure with subplots
fig, axs = plt.subplots(1, 2, figsize=(14, 7))

# Pie chart for Sales
axs[0].pie(category_df['Sales'], labels=category_df['Category'], autopct='%1.1f%%', startangle=90, colors=['#ff9999','#66b3ff','#99ff99'])
axs[0].set_title('Cumulative Sales per Category')

# Pie chart for Gross Profit
axs[1].pie(category_df['Profit'], labels=category_df['Category'], autopct='%1.1f%%', startangle=90, colors=['#ff9999','#66b3ff','#99ff99'])
axs[1].set_title('Cumulative Profit per Category')

# Adding a main title
plt.suptitle('Cumulative Sales and Gross Profit Distribution by Category', fontsize=16)

# Display the plot
plt.show()
```

```python
# Plotting the cumulative sales and gross profit per category on the same graph

# Define the position of the bars
bar_width = 0.35
r1 = range(len(category_df['Category']))
r2 = [x + bar_width for x in r1]

# Create the figure and axis
fig, ax = plt.subplots(figsize=(10, 6))

# Plotting sales
ax.bar(r1, category_df['Sales'], color='b', width=bar_width, edgecolor='grey', label='Sales')

# Plotting gross profit
ax.bar(r2, category_df['Gross'], color='r', width=bar_width, edgecolor='grey', label='Gross Income')

# Adding labels and title
ax.set_xlabel('Category', fontweight='bold')
ax.set_ylabel('Amount', fontweight='bold')
ax.set_title('Cumulative Sales and Gross Profit per Category', fontweight='bold')
ax.set_xticks([r + bar_width / 2 for r in range(len(category_df['Category']))])
ax.set_xticklabels(category_df['Category'])

# Adding a legend
ax.legend()

# Display the plot
plt.tight_layout()
plt.show()
```
