## Reading CSV file

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("superstore.csv")
```

```python
# Getting unique categoreis, sub categories and product names
categories = df["Category"].unique()

print(categories)
```

```python
# Finding the cumliative sales and profit per category
cat_1 = df[df["Category"]==categories[0]]
cat_2 = df[df["Category"]==categories[1]]
cat_3 = df[df["Category"]==categories[2]]

all_cats = [cat_1,cat_2,cat_3] 
sales = []
profits = []

for x in all_cats:
    total_sales = x["Sales"].sum()
    sales.append(total_sales)
    
    total_profit = x["Profit"].sum()
    profits.append(total_profit)    

# Creating a new table wth just categories
category_df = pd.DataFrame({
    "Category": categories,
    "Sales":sales,
    "Profit":profits
})

category_df
```

```python
# For sub category of Technology
sub_cat_3 = cat_3["Sub-Category"].unique()
all_sub_cats = [] 
tech_sub_sales = []
tech_sub_profits = []

# Looping through sub cats of Technology
for i in sub_cat_3:
    temp = cat_3[cat_3["Sub-Category"] == i]
    total_sales = temp["Sales"].sum()
    tech_sub_sales.append(total_sales)
    
    total_profit = temp["Profit"].sum()
    tech_sub_profits.append(total_profit)   

# Creating a new table wth just sub categories of technology
tech_df = pd.DataFrame({
    "Category": sub_cat_3,
    "Sales":tech_sub_sales,
    "Profit":tech_sub_profits
})

tech_df.sort_values(by='Profit', ascending=False)
```
